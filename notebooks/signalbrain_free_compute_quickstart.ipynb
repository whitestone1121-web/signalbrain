```python
import json, subprocess, sys, tempfile
from pathlib import Path

root = Path(tempfile.mkdtemp(prefix='signalbrain-notebook-'))
ledger = root / 'ledger.jsonl'
rows = [
    {
        'receipt_id': f'00{i:02d}-tooling-demo',
        'claim': f'00{i:02d}-tooling-demo',
        'confidence': 0.9,
        'held': True,
        'scored_by': 'measured',
        'change_class': 'tooling',
        'verdict': 'improvement',
        'measure_errors': [],
    }
    for i in range(10)
]
ledger.write_text(''.join(json.dumps(row) + '\n' for row in rows), encoding='utf-8')
result = subprocess.run(
    [sys.executable, '-m', 'signalbrain.cli', 'gate', '--ledger', str(ledger), '--by-class'],
    capture_output=True,
    text=True,
)
print(result.stdout)
print(result.stderr)
assert result.returncode == 0
```
